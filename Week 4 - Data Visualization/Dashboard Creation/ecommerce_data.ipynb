```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df_ec = pd.read_csv("ecommerce.csv")
df_ec
```

```python
df_ec.head(10)
```

```python
df_count = df_ec.groupby(['category']).count()
df_count.plot(kind = 'pie', y='price', autopct = '%1.0f%%')
plt.title("Category wise sales")
plt.show()
```

```python
df_ec[['category']].value_counts().plot(kind = 'bar')
```

```python
df_quarters = df_ec[df_ec['item'] == 'Mobile'].groupby(['quarter']).sum()
df_quarters[['price']].plot(kind='barh')
plt.title("Mobile sales per quarter")
plt.show()
```

```python
df_ec[['quarter']].value_counts().plot(kind = 'bar')
```

```python
import seaborn as sns
df_month = df_ec.groupby(['month']).sum()
df_month = df_month.reset_index()
sns.scatterplot(x = 'month', y = 'price', data=df_month, hue = 'month', legend=None)
sns.lineplot(x = 'month', y = 'price', data = df_month)
plt.title("Month wise sales")
```
